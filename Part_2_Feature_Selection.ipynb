#5. Feature engineering

```r
#Checking our sales data 
#

glimpse(Sales)
```

```r
#One-hot encoding of categorical variables
#

install.packages('dummies')
library(dummies)
Sales_final <- dummy.data.frame(Sales, names=c("Branch","Customer.type","Gender","Product.line",
                "Payment"), sep="_")
head(Sales_final)
```

```r
#First checking out the class of each column

class <- sapply(Sales_final, class)
class

#Selecting only numerical variables
#

Sales_final_Num <- Sales_final %>% select(which(class != "character"))
head(Sales_final_Num)
```

#6. Feature Selection

##6.a) Filter methods

```r
# Installing and loading our caret package
# 

suppressWarnings(
        suppressMessages(if
                         (!require(caret, quietly=TRUE))
                install.packages("caret")))
library(caret)
```

```r
#Installing and loading the corrplot package for plotting
#
 
suppressWarnings(
        suppressMessages(if
                         (!require(corrplot, quietly=TRUE))
                install.packages("corrplot")))
library(corrplot)
```

```r
# Calculating the correlation matrix of only continuous variables
#

correlationMatrix <- cor(Continuous_Variables)
correlationMatrix
```

```r
#Find attributes that are highly correlated
#

highlyCorrelated <- findCorrelation(correlationMatrix, cutoff=0.75)
highlyCorrelated

#Displaying the names of the hoghly correlated continuous variables
names(Continuous_Variables[,highlyCorrelated])

#The variables that are highly correlated are Tax,Cogs and Gross Income.
```

```r
# Removing Redundant Features 
#

Continuous_Variables2 <- Continuous_Variables[-highlyCorrelated]
head(Continuous_Variables2)
```

```r
# Performing our graphical comparison
# 

par(mfrow = c(1, 2))
corrplot(correlationMatrix, order = "hclust")
corrplot(cor(Continuous_Variables2), order = "hclust")


#From the below comparison you can note that the continuous variables remaining
#do not have a relatively perfect positive correlation like before.
```

##6.b) Wrapper Methods 

```r
# Installing and loading our clustvarsel package
# 
suppressWarnings(
        suppressMessages(if
                         (!require(clustvarsel, quietly=TRUE))
                install.packages("clustvarsel")))
                         
library(clustvarsel)
```

```r
#Installing and loading our mclust package
#

suppressWarnings(
        suppressMessages(if
                         (!require(mclust, quietly=TRUE))
                install.packages("mclust")))
library(mclust)
```

```r
# Sequential forward greedy search (default)
#

out = clustvarsel(Continuous_Variables, G = 1:5)
out

#Only two items are selected as significant,that is Tax and Unit Price
```

```r
#Having identified the variables that we use, we proceed to build the clustering model 
#

Subset1 = Sales_final_Num[,out$subset]
mod = Mclust(Subset1, G = 1:5)
summary(mod)
```

```r
#Plotting the model
#

plot(mod,c("classification"))
```

##6.d) Feature Ranking

```r
#Installation and loading of the required packages
# 

suppressWarnings(
        suppressMessages(if
                         (!require(FSelector, quietly=TRUE))
                install.packages("FSelector")))
library(FSelector)
```

```r
#Using of entropy - based approach to perform feature ranking
#

Scores <- information.gain(Total~., Sales_final_Num)
Scores
```

```r
#Algorithm to select a cut-off subset
#

Subset <- cutoff.k(Scores, 5)
as.data.frame(Subset)
```

```r
#Setting cutoff as a percentage
# that we would want to work with the percentage of the best variables.

Subset2 <-cutoff.k.percent(Scores, 0.2)
as.data.frame(Subset2)
```

#7. Insights from Feature Selection

We were able to perform feature selection on our dataset using three main techniques that is Filter methods, wrapper methods and feature ranking. Filter methods led to rating,unit price and total as most important variables. Wrapper methods selected most impactful variables as tax and unit price. The last method of feature ranking selected tax, cogs, gross income,quantity and unit price variables.

All methods of feature selection lead to selection of unique variables as the most important ones.
